```python
# Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on ‘Area’.  
# Which of these Areas had the 7th lowest sum in 2017?

grouped = df.groupby('Area')['Y2017'].sum().reset_index()
sorted_grouped = grouped.sort_values(by ='Y2017',ascending = True)

seventh_lowest_area = sorted_grouped.iloc[6]

print(seventh_lowest_area)
```

```python
# Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on ‘Area’.  Which of these Areas had the highest sum in 2017?

grouped = df.groupby('Area')['Y2017'].sum().reset_index()
highest_sum_area = grouped[grouped['Y2017'] ==grouped['Y2017'].max()]
print(highest_sum_area)
```
